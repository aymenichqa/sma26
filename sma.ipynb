```python
import csv

enc = tiktoken.get_encoding("cl100k_base")
vocab = enc._mergeable_ranks

print("Taille du vocabulaire :", len(vocab))

for i, (token_bytes, token_id) in enumerate(vocab.items()):
    if i < 100:
        print(token_bytes, token_id)

with open("mergeable_ranks.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, quoting=csv.QUOTE_ALL, escapechar="\\")
    writer.writerow(["token_bytes", "decoded", "rank"])
    for token, rank in vocab.items():
        token_bytes = repr(token)
        try:
            decoded = token.decode("utf-8")
        except UnicodeDecodeError:
            decoded = ""
        decoded = decoded.replace("\n", "\\n").replace("\r", "\\r")
        writer.writerow([token_bytes, decoded, rank])

print("Fichier mergeable_ranks.csv exporté.")
```

**1. Pourquoi faut-il créer `.gitignore` contenant `.env` avant le premier commit Git ?**

Car Git versionne tout ce qui est ajouté, y compris l'historique. Si `.env` est commité même une seconde, la clé API est dans l'historique Git — même après suppression du fichier, elle reste accessible. Créer `.gitignore` en premier empêche `.env` d'être jamais suivi.

**2. Quelle différence entre `tiktoken.encoding_for_model()` et `tiktoken.get_encoding()` ?**

`encoding_for_model("gpt-4o")` est un raccourci : il consulte une table interne pour trouver quel encodage utilise ce modèle (`o200k_base`), puis charge cet encodage. `get_encoding("o200k_base")` charge directement l'encodage par son nom. Les deux produisent le même objet tokenizer.

**3. Pourquoi le même mot peut avoir un token ID différent selon l'encodage ?**

Chaque encodage (cl100k_base, o200k_base) possède son propre vocabulaire construit indépendamment lors de l'entraînement du modèle. Les mots sont les mêmes, mais les identifiants numériques (IDs) assignés à chaque token sont arbitraires et propres à chaque vocabulaire.

**4. Dans LangChain, quel est le rôle du paramètre `temperature=0` ?**

`temperature` contrôle le degré d'aléatoire de la réponse. À `0`, le modèle choisit toujours le token le plus probable → réponse **déterministe et cohérente**, idéale pour des tâches techniques. Plus la valeur est haute (max 2), plus les réponses sont créatives mais moins prévisibles.

**5. Avantage et inconvénient d'un LLM local (Ollama) vs cloud (OpenAI) ?**

- **Ollama (local)** : ✅ Gratuit, 🔒 confidentiel (données ne quittent pas la machine) | ❌ Plus lent, qualité variable selon la taille du modèle
- **OpenAI (cloud)** : ✅ Très performant, rapide, modèles à la pointe | ❌ Payant, données envoyées à un serveur externe

**6. Qu'est-ce qu'un Agent IA ? En quoi diffère-t-il d'un simple appel à un LLM ?**

Un Agent IA est un système capable de **percevoir son environnement, planifier, prendre des décisions et agir** de manière autonome pour atteindre un objectif. Contrairement à un simple appel LLM (question → réponse unique), un agent peut enchaîner plusieurs appels, utiliser des outils (recherche web, code, APIs), s'adapter aux résultats intermédiaires et boucler jusqu'à l'accomplissement de la tâche.

**7. À quoi sert `sys.executable` dans une cellule Jupyter ?**

`sys.executable` retourne le chemin complet de l'interpréteur Python utilisé par Jupyter. C'est utile dans un environnement virtuel (`uv`) pour s'assurer que Jupyter utilise bien le Python de `.venv` et non un Python global. Si le chemin ne contient pas `.venv`, les paquets installés avec `uv add` ne seront pas trouvés.

**8. Pourquoi utilise-t-on `latin1` plutôt que UTF-8 pour décoder certains bytes tiktoken ?**

UTF-8 est un encodage à taille variable : certaines séquences de bytes (ex: `0xa1` seul) sont invalides. `latin1` (ISO-8859-1) mappe chaque byte `0x00–0xFF` vers exactement un caractère Unicode, donc **n'échoue jamais**. C'est idéal comme fallback pour afficher des bytes bruts du vocabulaire qui ne forment pas de séquences UTF-8 valides.

**9. Différence entre `os.environ["X"]` et `os.getenv("X")` ?**

- `os.environ["X"]` : lève une `KeyError` si `X` n'existe pas → utile en **production** pour détecter immédiatement une configuration manquante
- `os.getenv("X")` : retourne `None` silencieusement si `X` absent → plus confortable en **développement**, permet de tester la présence avant utilisation

**10. Différence entre `SystemMessage`, `HumanMessage` et `AIMessage` ?**

- `SystemMessage` : instructions générales au modèle (personnalité, contexte, règles) — correspond au rôle `system`
- `HumanMessage` : message envoyé par l'utilisateur — correspond au rôle `user`
- `AIMessage` : réponse générée par le modèle — correspond au rôle `assistant`

**11. Qu'est-ce que `bind_tools()` ? Dans quel cas l'utilise-t-on ?**

`bind_tools()` retourne un nouveau LLM enrichi de capacités supplémentaires (outils). Le modèle décide lui-même quand utiliser un outil selon la demande. On l'utilise quand on veut donner au LLM des capacités étendues : génération d'images, recherche web, exécution de code, appel d'APIs externes. C'est la base de l'architecture agentique.

**12. Étapes pour envoyer une image locale à un LLM multimodal via LangChain ?**

1. Ouvrir le fichier en mode binaire (`open(path, 'rb')`)
2. Encoder en base64 (`base64.b64encode(file.read()).decode('utf-8')`)
3. Créer un `HumanMessage` avec deux blocs de contenu : `type: text` (la question) et `type: image_url` (le Data URI base64)
4. Envoyer via `llm.invoke([HumanMessage(...)])`
5. Afficher `resp.content` (texte) ou `display(Markdown(resp.content))`

**13. Pourquoi encoder l'image en base64 avant envoi au LLM ?**

Les APIs utilisent JSON comme format d'échange, et JSON ne supporte que du texte — pas des données binaires brutes. base64 résout ce problème en transformant les bytes de l'image en une chaîne de texte ASCII qui peut être incluse dans du JSON. C'est un encodage universel utilisé partout où des données binaires doivent transiter via des protocoles textuels.
