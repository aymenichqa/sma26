```python
enc1 = tiktoken.encoding_for_model("gpt-4o")   # o200k_base
enc2 = tiktoken.get_encoding("cl100k_base")     # GPT-4

text = "Hello"
tokens1 = enc1.encode(text)
tokens2 = enc2.encode(text)

print("Tokens gpt-4o :", tokens1)
print("Tokens cl100k :", tokens2)

for t in tokens1:
    print("Token1:", t, "->", enc1.decode([t]))
for t in tokens2:
    print("Token2:", t, "->", enc2.decode([t]))
```

```python
enc = tiktoken.encoding_for_model("gpt-4o")

tokens1 = enc.encode("Chat")
tokens2 = enc.encode("LeChat")

for t in tokens1:
    print("Chat token:", t, "->", enc.decode([t]))
for t in tokens2:
    print("LeChat token:", t, "->", enc.decode([t]))
```

```python
import csv

enc = tiktoken.get_encoding("cl100k_base")
vocab = enc._mergeable_ranks

print("Taille du vocabulaire :", len(vocab))

for i, (token_bytes, token_id) in enumerate(vocab.items()):
    if i < 100:
        print(token_bytes, token_id)

with open("mergeable_ranks.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, quoting=csv.QUOTE_ALL, escapechar="\\")
    writer.writerow(["token_bytes", "decoded", "rank"])
    for token, rank in vocab.items():
        token_bytes = repr(token)
        try:
            decoded = token.decode("utf-8")
        except UnicodeDecodeError:
            decoded = ""
        decoded = decoded.replace("\n", "\\n").replace("\r", "\\r")
        writer.writerow([token_bytes, decoded, rank])

print("Fichier mergeable_ranks.csv exporté.")
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv(override=True)

llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0,
    api_key=""  # lue depuis .env automatiquement
)

print(dir(llm))
```

```python
resp = llm.invoke(input=[
    {"role": "system", "content": "Vous etes un expert en IA Agentique"},
    {"role": "user",   "content": "C'est quoi un Agent IA"}
])

print(resp)          # Objet AIMessage complet
print("---")
print(resp.content)  # Texte seulement
```
